```python
import acoular as ac
import numpy as np
import matplotlib.pyplot as plt

grid = ac.RectGrid(x_min=-1, x_max=1, y_min=-1, y_max=1, z=1, increment=1)

sector = ac.RectSector(x_min=-0.2, x_max=0.2, y_min=-0.2, y_max=0.2)

sd = grid.subdomain(sector)
print("RECT:")
print("size:", grid.size)
print("pos: ", np.shape(grid.pos))
print("subd: ", np.shape(sd))
print(grid.subdomain(sector)[0].shape[0])
print(grid.pos[sd])

XS = np.linspace(grid.x_min, grid.x_max, 1000)
YS = np.linspace(grid.x_min, grid.x_max, 1000)
ZS = np.ones_like(XS)[..., np.newaxis]

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(*grid.pos, "o", label="rect grid")
# ax.plot_surface(*(XS, YS, ZS), label="rect subd")


grid_polar = ac.PolarGrid(r_min=0., r_max=2, theta_min=0., theta_max=360., z=1., r_increment=1, theta_increment=90.)
sd_polar = grid_polar.subdomain(sector)
print("\nPOLAR:")
print("size:", grid_polar.size)
print("pos: ", np.shape(grid_polar.pos))
print("subd", np.shape(sd_polar))
print(grid_polar.subdomain(sector)[0].shape[0])
print(grid_polar.pos[sd_polar])
ax.plot(*grid_polar.pos, "o", label="polar grid")

plt.legend()
```
